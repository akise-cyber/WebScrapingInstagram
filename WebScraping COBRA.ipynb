```python
!pip install selenium
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait # available since 2.4.0
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time


#colocar o caminho até o chromedriver.exe

driver = selenium.webdriver.Chrome(executable_path=r"C:\Users\carli\Downloads\webscraping\chromedriver_win32/chromedriver.exe")

#abre o instagram
driver.get("http://www.instagram.com")

username = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']")))
password = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']")))

#seu usuário e senha
username.clear()
username.send_keys("seu login")
password.clear()
password.send_keys("sua senha")

#clicka em entrar
button = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
```

```python
time.sleep(5)
alert = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Agora não")]'))).click()
alert2 = WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Agora não")]'))).click()
```

```python
#procura a barra de pesquisa
searchbox = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Pesquisar']")))
searchbox.clear()

#procura a hashtag
keyword = "#cobra"
searchbox.send_keys(keyword)
 
time.sleep(5) # Wait for 5 seconds
my_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[contains(@href, '/" + keyword[1:] + "/')]")))
my_link.click()
```

```python
n_scrolls = 1
for j in range(0, n_scrolls):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
```

```python
#os elementos da página
anchors = driver.find_elements_by_tag_name('a')
anchors = [a.get_attribute('href') for a in anchors]
#pega apenas os com links de imagem
anchors = [a for a in anchors if str(a).startswith("https://www.instagram.com/p/")]

print('Encontrei ' + str(len(anchors)) + ' links de imagens')
anchors[:5]
```

```python
images = []

#extrai as imagens
for a in anchors:
    driver.get(a)
    time.sleep(2)
    img = driver.find_elements_by_tag_name('img')
    img = [i.get_attribute('src') for i in img]
    images.append(img[1])
    
images[:5]
```

```python
import os
!pip install wget
import wget

path = os.getcwd()
path = os.path.join(path, keyword[1:] + "s")

#cria a pasta com o nome da hashtag
os.mkdir(path)
path
```

```python
#baixa todas as imagens para a pasta
counter = 0
for image in images:
    save_as = os.path.join(path, keyword[1:] + str(counter) + '.jpg')
    wget.download(image, save_as)
    counter += 1
```
